```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample confusion matrix data
confusion_matrix = np.array([[524, 10, 8],
                            [15, 307, 8],
                            [8, 10, 281]])

def get_classification_report(confusion_matrix):
  """
  Calculates classification report metrics from a confusion matrix.

  Args:
    confusion_matrix: A 2D numpy array representing the confusion matrix.

  Returns:
    A dictionary containing the following classification metrics:
      - accuracy
      - precision
      - recall
      - f1_score
  """

  # Extract values from the confusion matrix
  tp = confusion_matrix[0, 0]
  fn = confusion_matrix[1, 0]
  fp = confusion_matrix[0, 1]
  tn = confusion_matrix[1, 1]

  # Calculate metrics
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  precision = tp / (tp + fp)
  recall = tp / (tp + fn)
  f1_score = 2 * (precision * recall) / (precision + recall)

  # Create a dictionary to store the results
  report = {
      'accuracy': accuracy,
      'precision': precision,
      'recall': recall,
      'f1_score': f1_score
  }

  return report

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Oranges", 
            xticklabels=["BP", "NL", "VP"],
            yticklabels=["BP", "NL", "VP"], annot_kws={"fontsize":14})

# Calculate classification report
report = get_classification_report(confusion_matrix)

# Print the classification report
print("Classification Report:")
print(f"Accuracy: {report['accuracy']:.4f}")
print(f"Precision: {report['precision']:.4f}")
print(f"Recall: {report['recall']:.4f}")
print(f"F1-Score: {report['f1_score']:.4f}")

# Customize x and y labels (optional)
plt.xlabel("Predicted Label", fontweight="bold")  # Make predicted label bold
plt.ylabel("True Label", fontweight="bold")  # Make true label bold
plt.title("Confusion Matrix",fontweight='bold')
plt.show()   # To display the plot
#plt.savefig("confusion_matrix.png")  # Save the plot as an image
```

```python

import cv2

def enhance_image(image_path):
    """Enhances an image using OpenCV for publication.

    Args:
        image_path (str): The path to the image file.

    Returns:
        numpy.ndarray: The enhanced image.
    """

    # Load the image
    image = cv2.imread(image_path)

    # Convert to grayscale for preprocessing
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise, but use a smaller kernel for less blurring
    blurred = cv2.GaussianBlur(gray, (3, 3), 0)

    # Enhance contrast using histogram equalization, but adjust the clip limits to prevent over-enhancement
    equalized = cv2.equalizeHist(blurred)

    # Apply adaptive thresholding with a larger block size for better edge preservation
    thresh = cv2.adaptiveThreshold(equalized, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)

    # Apply morphological operations for noise reduction and edge enhancement, but with fewer iterations
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=1)
    eroded = cv2.erode(closed, kernel, iterations=1)
    dilated = cv2.dilate(eroded, kernel, iterations=1)

    # Combine enhanced image with original for better results, but adjust the alpha value to control the level of enhancement
    enhanced = cv2.addWeighted(image, 1.2, image, 0, 0, dtype=cv2.CV_32F)
    enhanced = cv2.convertScaleAbs(enhanced)

    return enhanced

# Example usage
image_path = "/home/idrone2/Desktop/rk/penumonia/cnn-graph.png"
enhanced_image = enhance_image(image_path)

# Save the enhanced image
cv2.imwrite("/home/idrone2/Desktop/rk/penumonia/enhanced_cnn-graph.jpg", enhanced_image)
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic image data (150x150 grayscale images, 3 classes)
def generate_synthetic_data(num_samples=1500):
    # Random images (150x150 grayscale)
    X = np.random.rand(num_samples, 150, 150, 1)  # Grayscale 150x150 images
    
    # Labels for 3 classes: 0 (BP), 1 (N), 2 (VP)
    y = np.random.choice([0, 1, 2], num_samples, p=[0.6, 0.25, 0.15])  # Imbalanced data
    return X, y

# Generate data
X, y = generate_synthetic_data()

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert labels to one-hot encoding
y_train_one_hot = to_categorical(y_train, 3)
y_test_one_hot = to_categorical(y_test, 3)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Build a simple CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(3, activation='softmax')  # 3 output classes (BP, N, VP)
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_one_hot, epochs=5, batch_size=32, validation_data=(X_test, y_test_one_hot))


from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict class probabilities
y_pred_probs = model.predict(X_test)

# Convert predicted probabilities to class labels
y_pred_classes = np.argmax(y_pred_probs, axis=1)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_classes)
print("Confusion Matrix:")
print(cm)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', cbar=True, xticklabels=['BP', 'N', 'VP'], yticklabels=['BP', 'N', 'VP'])
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()


from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Binarize the test labels for ROC curve calculation
y_test_binarized = label_binarize(y_test, classes=[0, 1, 2])

# Dictionary to store the False Positive Rate (FPR), True Positive Rate (TPR), and AUC for each class
fpr = {}
tpr = {}
roc_auc = {}

# Compute ROC curve and AUC for each class
for i in range(3):
    fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

    # Plot ROC curve for each class
    plt.figure()
    plt.plot(fpr[i], tpr[i], color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for Class {i} (BP=0, N=1, VP=2)')
    plt.legend(loc="lower right")
    plt.show()

# Print the AUC for all classes
for i in range(3):
    print(f"AUC for class {i}: {roc_auc[i]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_roc_curve(confusion_matrix):
    """Calculates the ROC curve and AUC from a confusion matrix.

    Args:
        confusion_matrix: A 2x2 confusion matrix.

    Returns:
        fpr: False Positive Rate.
        tpr: True Positive Rate.
        auc: Area Under the Curve (AUC).
    """

    tp = confusion_matrix[1, 1]
    tn = confusion_matrix[0, 0]
    fp = confusion_matrix[0, 1]
    fn = confusion_matrix[1, 0]

    tpr = tp / (tp + fn)
    fpr = fp / (fp + tn)

    roc_curve = np.array([[0, 0], [fpr, tpr]])
    auc = np.trapz(roc_curve[:, 1], roc_curve[:, 0])

    return fpr, tpr, auc

# Example usage:
confusion_matrix = np.array([[140, 2, 0],
                            [0, 328, 2],
                            [2, 0, 297]])

fpr, tpr, auc = calculate_roc_curve(confusion_matrix)

# Plot the ROC curve
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()

print("AUC:", auc)
```
